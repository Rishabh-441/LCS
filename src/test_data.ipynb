```python
import pandas as pd
df = pd.read_csv('../DataCreation/merged_processed_output.csv')
df.head()
```

```python
import ast
doc = []
summ = []
for i in range(0, 10) :
    try :
        doc_text = ''.join(ast.literal_eval(df['document'].iloc[i]))
        summ_text = ''.join(ast.literal_eval(df['summary'].iloc[i]))
        doc.append(doc_text)
        summ.append(summ_text)
    except Exception :
        continue

print(df['document'].iloc[0])
print(doc[0])
print(summ)        
```

```python
import pandas as pd

data = {
    'document': doc,
    'summary': summ
}


new_df = pd.DataFrame(data)

new_df.to_csv('generated_data/legal_summaries.csv', index=False)

print(f"Successfully saved {len(new_df)} documents and summaries to 'legal_summaries.csv'")
```
